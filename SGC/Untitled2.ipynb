```python
def get_response(json_file):
    import requests
    import json
    with open(json_file) as f:
        para = json.load(f)

    headers = {
        'Accept': 'application/json',
    }

    data = {
      'key': '20ab9d1d-ff00-4cef-a51f-e466441240f6',
      'parameters': json.dumps(para)
    }

    response = requests.post('https://www.glastuinbouwmodellen.wur.nl/kaspro/TenCent/', headers=headers, data=data, verify=False)
    res_dict = json.loads(response.text)
    return res_dict
```

```python
res_dict = get_response('settings.json')
```
